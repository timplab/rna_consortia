### Debruijn data analysis (GGACU)

data is located in "/dilithium/Data/Nanopore/rna/debruijn" 

below is my code to extract the event means from 3 positions in the fasta file that contain GGACU and plot the distribution at each position

```python
from IPython.display import Image
Image(filename='debruijn.png') 
```

```python
debruijn_file = open('debruijn.eventalign.txt')

deb_190_GGACU = []
deb_356_GGACU = []
deb_453_GGACU = []

for line in debruijn_file:
    line_arr = line.strip().split('\t')
    pos = line_arr[1]
    event_mean = line_arr[6]
    model_kmer = line_arr[9]
    ref_kmer = line_arr[2]
    if model_kmer == 'GGACT' and ref_kmer == 'GGACT':
        
        if pos == '190':
            deb_190_GGACU.append(float(event_mean))
            
        if pos == '356':
            deb_356_GGACU.append(float(event_mean))
            
        if pos == '453':
            deb_453_GGACU.append(float(event_mean))
            
debruijn_file.close()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
sns.set(color_codes=True)
%matplotlib inline

params = {'legend.fontsize': 30,
          'legend.handlelength': 2,
          'axes.titlesize' : 40,
          'axes.labelsize' : 40,
          'lines.linewidth' : 3,
          'lines.markersize' : 10,
          'xtick.labelsize' : 25,
          'ytick.labelsize' : 25}
plt.rcParams.update(params)
sns.set_style('ticks')
plt.rcParams['figure.figsize']=30,10



sns.kdeplot(deb_190_GGACU, shade=True, label= 'deb_190_GGACU')
sns.kdeplot(deb_356_GGACU, shade=True, label= 'deb_356_GGACU')
sns.kdeplot(deb_453_GGACU, shade=True, label= 'deb_453_GGACU')


plt.axvline(x=123.8 ,color='k', linestyle='--' )
plt.ylim(0, 0.2)
plt.xlim(90, 150)
plt.title( 'GGACU at 3 different positions in debruijn data')
plt.xlabel('Event_level_mean')
plt.ylabel('Density')
```
