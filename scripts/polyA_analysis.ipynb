```python
isoform_file = open('./nvrna.180802.read.isoform.map.txt')
polya_file = open('NA12878_DirectRNA_polyA_primary.txt')
gene_to_isoforms = {}
read_to_isoform = {}
isoform_to_polya_length = {}
polya_lengths = []
for line in isoform_file:
    
    read_id = line.split('\t')[0]
    isoform_id = line.split('\t')[1].split('_')[0]
    if isoform_id not in gene_to_isoforms.setdefault(line.strip().split('\t')[1].split('_')[1].split('.')[0],[]):
        
        gene_to_isoforms.setdefault(line.strip().split('\t')[1].split('_')[1].split('.')[0],[]).append(isoform_id)
#     read_to_isoform.setdefault(read_id,[]).append(isoform_id)
    read_to_isoform[read_id] = isoform_id
    
    
for line in polya_file:
    line_arr = line.strip().split('\t')
    read_id = line_arr[0]
    polya_length = line_arr[8]
    qc_taq = line_arr[9]
    if qc_taq == 'PASS' and read_id in read_to_isoform.keys():
        
        isoform_to_polya_length.setdefault(read_to_isoform[read_id],[]).append(float(polya_length))
        polya_lengths.append(float(polya_length))
        
        

                             
isoform_file.close()
polya_file.close()
```

```python
sns.set_style("darkgrid", {"axes.facecolor": ".9"})
def kde(gene_list, gene_to_lengths, gene_name):
    colors = ['b','g','r','purple','yellow']
    arr = []
    for i in gene_list:
        arr.append(list(gene_to_lengths[i]))
        
    
        
    plt.figure()
    for ii in range(len(arr)):
        
        sns.kdeplot(arr[ii], shade=True, label= gene_name[ii], color=colors[ii])
#     ax = sns.swarmplot(data=arr, color="white", edgecolor="gray" )
#     ax.set_xticklabels(gene_list)
    plt.title(  '  PolyA length analysis for some genes with >20000 reads spanning each gene' )
    plt.xlabel('polyA length')
    plt.ylabel('Density')
#     plt.show()
```
